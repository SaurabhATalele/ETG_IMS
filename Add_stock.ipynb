```python
#This program adds the stock to the inventory and also the batch details for further reference


import json
import datetime

stock = {}

batch_no = int(input("Enter Batch Number:"))

#Getting previous inventory data

file = open("Data/Products.json","r")

data = file.read()
data = json.loads(data)

file.close()


#taking stock input

while True:
    pid = input("Enter Product ID: ")
    
    if pid not in data.keys():
        dec = input("The Product Does not Exists in the inventory! Do you Want to add it? (y to add, n to Try Again) :")
        if dec == "y":
            break
        else:
            continue
    else:
        quantity = int(input("Enter the quantity: "))
    
    data[pid]["Quantity"] += quantity 
    
    if pid in stock.keys():
        stock[pid] += quantity
    else:
        stock[pid] = quantity
    
    inp = input("Do you want to add more stock? (y for yes )")
    if not inp == "y":
        break

        
#Updating inventory with the new stock data 

final_data = json.dumps(data)

file = open("Data/Products.json","w")

file.write(final_data)
file.close()


#Getting batch data

file = open("Data/Batches.json",'r')

data = file.read()
if data != "" :
        data = json.loads(data)
else:
    data = {}

file.close()


#Creating new batch data 

x = datetime.datetime.now()
data[batch_no] = {
    "date" : str(datetime.date.today()),
    "time" : str(x.strftime("%c")),
    "stock" : stock
    
}


#Updating The batch data with new batch
data = json.dumps(data)

file = open("data/Batches.json","w")

file.write(data)
file.close()
#Printing thanks message

print("\n\n*******************\nYour Record has been Saved!!!\nThank You!!!")
    
```
